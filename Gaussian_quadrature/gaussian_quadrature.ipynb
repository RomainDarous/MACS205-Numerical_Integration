# Gaussian quadrature : general case.

Let $\chi ⊂ ℝ$. Let $\omega$ be a non-negative weight function $\left(\text{i.e }\int_\chi \omega = 1 \right)$. For any polynomials $P, Q$ defined over $\chi$,  their scalar product for the weight function $\omega$ is defined by :

$$\left\langle{P},{Q}\right\rangle_\omega = \int_\chi P(x)Q(x)\omega(x)dx$$

Let $f$ be a continuous function over $\chi$.  We are interested in finding the optimal weights $(\omega_i)_i$ and nodes $(x_i)_i$ maximizing the order of the quadrature rule for the integral $I(f) = \int_\chi \omega f$ given by :

$$\widehat{I_n(f)} = \sum_{i=0}^{n}\omega_i f(x_i)$$

# **Part I. Theory**

**NOTE** : it is enough to show all the results for $(x_i)$ all differents.
Indeed, once it is proven in this case, we can do the general case as follows :

Suppose that $x_0 = x_1$, for instance:

$$\begin{align} \widehat{I_n(f)} & = (\omega_0 + \omega_1 ) f(x_0) + \sum_{i=2}^{n}\omega_i f(x_i)\\
&= \sum_{i=1}^{n-1}\omega_i' f(x_i) \\
\end{align}$$

where $\omega_0'=\omega_0+\omega_1$ and $\omega_i' = \omega_i$ for $i \geq 1$.
And we do this until all the nodes are different.

**Orthogonal Polynomials**.
Consider the orthogonal family of polynomials $(P_0,P_1,...)$ obtained by applying the Gram-Schmidt process to the basis $1, X, X^2, . . .$ with the inner product $\left\langle{},{}\right\rangle_\omega$ and the normalization condition $P_n(1) = 1$.

First, we prove an important property of any orthogonal family of
polynomials : each element $P_i$ has $i$ distinct simple roots in $\chi$.

   

1. Show that $P_n$ changes sign at least once in $\chi$

We know that $(P_0,P_1,...,P_n)$ is an orthogonal family. 

So $\left\langle{P_n},{P_0}\right\rangle_\omega = 0$. And we have $P_0 = 1$.

Finaly, $$\int_\chi P_n(x)\omega(x)dx = 0$$

Suppose that $P_n$ has a constant sign. Then, $P_n\omega$ has a constant sign as $\omega$ is a non-negative function.

Moreover, $$\int_\chi P_n(x)\omega(x)dx = 0$$

gives that $P_n\omega = 0$ on $\mathcal{X}$.

Let's denote $\mathcal{A} = \{ x \in \mathcal{X}, \omega(x) \neq 0\}$.

$$\int_\chi \omega(x)dx = \int_\mathcal{A} \omega(x)dx = 1$$

This gives that the measure of $\mathcal{A}$ is non zero (otherwise, one would have $\int_\mathcal{A}\omega = 0$) and:

$$\forall x \in \mathcal{A}, P_n(x)\omega(x) = 0 \Rightarrow P_n(x) = 0$$

Hence, $P_n$ has an infinite number of roots so $P_n = 0$. This is absurd as the Gram-Schmidt method gives that $\mathrm{deg}(P_n)=n$.

One concludes that $P_n$ changes sign at least once.

2. Let $y_1, . . . , y_j$ the only points in $\chi$ where $P_n$ changes its sign. Show that $j = n$ and conclude.

Suppose that $j < n$ :

We know that $(P_0,...,P_j)$ is a basis of $ℝ_j[X]$ and $\forall i \in \{0,...,j\}, \left\langle{P_n},{P_i}\right\rangle = 0$.

The points where $P_n$ changes sign are the ones with odd multiplicity.

So, one can denote $P_n = Q\prod_{i=1}^j(X-y_i)^{2\lambda_i +1}$ where $y_i$ is not a root of Q. So Q has a constant sign, as it has no root with odd multiplicity.

Let's note that: <br>
$\left\langle{P_n},{(X-y_1)...(X-y_j)}\right\rangle_\omega = 0$ as $\mathrm{deg}\left[(X-y_1)...(X-y_j)\right] = j < n$.

Then,

$$\int_\chi Q(x)\prod_{i=1}^j(x-y_i)^{2\lambda_i +2}\omega(x)dx = 0$$

What's inside the integral has a constant sign, and by using the same argument as in 1), one can show that ${P_n}{(X-y_1)...(X-y_j)} = 0$, which absurd.

So $j=n$. $P_n$ has n roots with odd multiplicity and $\mathrm{deg}(P_n)= n$. Hence, $P_n$ has n distinct roots.

Now we move on to show that the maximum order of the method is $2n + 1$ and that the nodes are unique and given by the roots of the orthogonal polynomial basis.

**Necessary conditions**. Assume the rule is of order $2n + 1$.

3. Using a particular choice for $f$, show that the nodes $(x_i)_i$ must be roots of $P_{n+1}$.

Let's start by giving an explicit form for the $(w_i)$ coefficients.

Let's denote, $\forall i \in {0,...,n}, L_i$ the lagrange polynomials for the points $(x_i)_{i \in \{0,...,n\}}$. 
<br><br>

Let $i \in \{0,...,n\}$.
$\mathrm{deg}(L_i) = n$ and $\mathrm{deg}(L_i^2) = 2n$.

As the rule is of order $2n+1$,

$$\widehat{I_n(L_i^2)} = I(L_i^2) = ||L_i||_\omega^2 > 0$$
as all the $x_i$ are different.

Moreover, 

$$\widehat{I_n(L_i^2)} = \sum_{j=0}^{n}\omega_j L_i^2(x_j) = \omega_i$$

Hence, $\forall i \in \{0,...,n\}, \omega_i = ||L_i||_\omega^2 > 0$.

Now, we come back to the question.

Let $j \in \{0,...,n\}$.

Now, we consider : $$f = P_{n+1}L_j$$

$f$ is of degree $2n+1$ and $L_j$ is of degree $n$.

$L_j \in ℝ_n[X]$ so we can decompose it on $(P_0,...,P_n)$, which is a basis of $ℝ_n[X]$

Then $\left\langle{P_{n+1}},{L_j}\right\rangle_\omega = 0$ because $(P_0,...,P_{n+1})$ is orthogonal.

So $I(P_{n+1}L_j) = 0$

And $\sum_{i=0}^{n}\omega_i P_{n+1}(x_i)L_j(x_i) = P_{n+1}(x_j)\omega_j = 0$.

So $P_{n+1}(x_j) = 0$ as $w_j \neq 0$. 

Then $\forall i \in \{0,...,n\}, P_{n+1}(x_i) = 0$

**Conclusion : the nodes $(x_i)_i$ must be roots of $P_{n+1}$**

4. Conclude the uniqueness of $(x_i)_i$ and $(\omega_i)_i$.

$P_{n+1}$ has exactly $n+1$ roots (cf 1. and 2.) so we have the uniqueness of $(x_i)_i$.

**The $(x_i)_i$ are exactly the $n+1$ roots of $P_{n+1}$**

Let $j \in \{0,...,n\}$

Let $$L_j(X) = \frac{\prod_{i \neq j} (X - x_i)}{\prod_{i \neq j} (x_j - x_i)}$$ the polynomious Lagragnge interpolation on the $(x_i)_i$.

We have $\forall i \neq j, L_j(x_i) = 0$ and $L_j(x_j) = 1$.

So $$\sum_{i=0}^{n}\omega_i L_j(x_i) = \omega_j = \int_\chi L_j(x)\omega (x) dx$$  

**The $(\omega_i)_i$ are exactly $\int_\chi L_i(x)\omega (x) dx$ and are therefore unique**




**Sufficient conditions.**

5. Using Euclidean division, show that the necessary conditions are sufficient for the rule to be of order at least $2n + 1$.

Now we consider $(x_i)_i$ and $(\omega_i)_i$ satisfying the necessary conditions. 

Let $f = Q $ a polynomious of degree $\leq 2n+1$.

We can write $Q = AP_{n+1} + R$ using the Euclidean division of $Q$ by $P_{n+1}$. 

$P_{n+1}$ of degree $n+1$ so $R$ of degree $\leq n$ and A is of degree $\leq n$.

$$I(Q) = \int_\chi Q\omega = \int_\chi R\omega$$ 

$$\widehat{I_n(Q)} = \sum_{i=0}^{n}\omega_i Q(x_i) = \sum_{i=0}^{n}\omega_i R(x_i) $$

And $\sum_{i=0}^{n}\omega_i R(x_i) = \sum_{i=0}^{n} (\int_\chi L_i(x) \omega (x)dx )R(x_i) = \int_\chi (\sum_{i=0}^{n} L_i(x)R(x_i))\omega (x) dx$.

Using the Newton-Costes method **of order n**, $R(x) = \sum_{i=0}^{n} L_i(x)R(x_i)$.

So $$\int_\chi R\omega = \sum_{i=0}^{n}\omega_i R(x_i)$$

And therefore : $$I(Q) = \widehat{I_n(Q)}$$

**Conclusion : the rule to be of order at least $2n + 1$**

6. Using a specific choice of $f$ , show that the order of the rule cannot be larger than $2n + 1$.

Let $f = P_{n+1}^2$ of degree $2n+2$.

$I(f) = \int_\chi P_{n+1}^2\omega = ||P_{n+1}||^2 > 0$ because $P_{n+1} \neq 0$.

$\widehat{I_n(f)} = \sum_{i=0}^{n}\omega_i P_{n+1}^2(x_i) = 0 $ because  the $x_i$ are all roots of $P_{n+1}$.

So the rule is of order $< 2n+2$

**Conclusion : The rule is of order $2n+1$**

Examples Let $n=1$.
- For $\mathcal{X}=[-1,1]$ and $w(x)=\frac{1}{2},\left(P_0, P_1, \ldots\right)$ are called the Legendre polynomials.
- For $\mathcal{X}=\mathbb{R}$ and $w(x)=\frac{1}{\sqrt{2 \pi}} e^{-\frac{x^2}{2}} .\left(P_0, P_1, \ldots\right)$ are called the Hermite polynomials.
- For $\mathcal{X}=[0,+\infty]$ and $w(x)=e^{-x},\left(P_0, P_1, \ldots\right)$ are called the Laguerre polynomials.


7. Find $P_0, P_1$ and $P_2$. Deduce the corresponding Gauss-Legendre quadrature for $n=1$

We know (Gram-Schmidt) that :

$$P_k(X) = X^k - \sum_{i=0}^{k-1}\frac{\left\langle{X^i},{P_i}\right\rangle_\omega}{\left\langle{P_i},{P_i}\right\rangle_\omega}P_{k-1}$$

So $$P_0 = 1$$ 
$$P_1 = X$$
$$P_2 = \frac{3X^2 -1}{2}$$

So $x_0,x_1 = -\frac{1}{\sqrt{3}},\frac{1}{\sqrt{3}}$ (roots of $P_2$)

and $\omega_0,\omega_1 = 1, 1$ when using their explicit formula.

8. For $\mathcal{X}=[a, b]$, what change of variable should be applied to the nodes $x_i$ ?

```scipy.special.roots_legendre``` performs the approximation of $\int^1_{-1}f \approx \sum^{n}_{i=0}w_i^Lf(x_i)$.

One may notice that

$\int^1_{-1}f = 2\int^1_{-1} \frac{1}{2} f = 2\int^1_{-1} w f$ where
$w : x \mapsto \frac{1}{2}$. <br>
We have $w \geq 0$ and $\int^1_{-1} w = 1$.
<br><br>

This approximation is equivalent to the quadrature we studied, taking $\mathcal{X} = [-1, 1]$ and the function $w$ mentionned above.

Then, we have that $\int^1_{-1}f = 2\int^1_{-1} w f \approx \sum^{n}_{i=0}w_i^Lf(x_i) \Rightarrow \int^1_{-1} w f \approx \sum^{n}_{i=0}\frac{w_i^Lf}{2}(x_i)$.

So one can get from one method to the other by diving or multiplying the weights by $2$.

Thus, ```scipy.special.roots_legendre``` is a **particular gaussian quadrature method**. It has all the properties showed in the first part.

<br>

Now, let's comme back to the question. <br>
**Reminder** : ```scipy.special.roots_legendre``` gives the nodes and weights which allow to approximate $\int^1_{-1}f \approx \sum^{n}_{i=0}w_i^Lf(x_i)$.

<br>

We now determine the variable change that allows to deduce the nodes used to compute $\int^b_{a}f$.

Let's distinguish two cases : 

A) $a = b$
In this case, one can easily compute $\int^b_af$, as it is exactly equal to $0$.

B) $a \neq b$

Let's consider the variable change $u(x) = \frac{-2}{a-b}x + \frac{a+b}{a-b}$,  $\mathcal{C}^1$ diffeomorphism and $u(a) = -1, u(b) = 1$.

$\int_a^b f(x) \mathrm(dx) = \frac{b-a}{2}\int_{-1}^1f(\frac{b-a}{2}(u-\frac{a+b}{a-b})) \mathrm{du} \approx \sum^{n}_{i=0}w_i^L\frac{b-a}{2}f(\frac{b-a}{2}(x_i-\frac{a+b}{a-b}))$.

One needs to apply to the ```scipy.special.roots_legendre``` to the function $g : x \mapsto \frac{b-a}{2}f(\frac{b-a}{2}(x-\frac{a+b}{a-b}))$.


9. Implement a function def gauss_legendre_simple($f, a, b,$ nodes, weights) that approximates the integral of a function $f$ over $[a,b]$ using the Gaussian-Quadrature rule. Test your implementation on some analytically integrable function of your choice.



````python
def gauss_legendre_simple(f, a, b, nodes, weights) :
    if (a==b) : return 0
    I_n = 0
    for i in range(len(nodes)) :
        I_n += weights[i] * (b-a)/2 * f((b-a)/2 * (nodes[i] - (a+b)/(a-b)))
    return I_n
````

10. Implement a function def gauss_legendre($f, n, M, a, b$) that performs the composite rule of Gauss-Legendre over M sub-intervals of $[a, b]$.

````python
from scipy.special import roots_legendre
import numpy as np

def gauss_legendre(f, n, M, a, b) :
    if (a==b) : return 0
    nodes, weights = roots_legendre(n+1) #one gets n nodes and weights, so the rule is of order 2n-1
    step = (b-a)/M
    I_n = 0
    a_values = np.arange(a, b, step)
    for a_0 in a_values:
        b_0 = a_0 + step
        I_n += gauss_legendre_simple(f, a_0, b_0, nodes, weights)
    return I_n
````

11. Write an identical function but with a vectorized implementation gauss_legendre_vectorized($f, \mathrm{n}, \mathrm{M}, \mathrm{a}, \mathrm{b}$ ) with numpy (does not contain any loop or list comprehensions. Hint : it should not call gauss_legendre_simple).


One uses the formula
$\int_a^b f(x) \mathrm(dx)  \approx \sum_{k=0}^M \sum^{n}_{i=0}w_i^L\frac{a_{k+1}-a_{k}}{2}f(\frac{a_{k+1}-a_{k}}{2}x_i + \frac{a_{k}+a_{k+1}}{2})$.

````python
from scipy.special import roots_legendre
import numpy as np

# f has to be able to take np.array argument
def gauss_legendre_vectorized(f, n, M, a, b):
    if(a==b) : return 0

    nodes, weights = roots_legendre(n+1)
    nodes = nodes.reshape(1, -1)
    weights = weights.reshape(-1,1)

    step = (b - a) / M
    
    a_np = np.arange(a, b, step)
    b_np = a_np + step
    diag_b_a = np.diag((b_np-a_np)/2)
    a_np = a_np.reshape(-1, 1)
    b_np = b_np.reshape(-1,1)

    res = diag_b_a @ f((b_np-a_np)/2 @ nodes + (( (a_np+b_np) / 2 )) @ np.ones((1,n+1))) # a row = all terms to sum of one subdivision without weight
    res = res @ weights # applying weight
    res = np.sum(res) # summing all results
    return res
````

````python
def trapezoide_rule(f, M, a, b):

  h = (b - a)/ M
  node = np.array( [a + i*h for i in range(M+1)])
  weight = np.repeat(1.0,M+1)
  weight[0] = 1/2
  weight[-1] = 1/2

  return h*np.sum(np.dot(f(node),weight))
````

$$\begin{align}

# **Part III : Richardson acceleration and Romberg’s method**

Consider an approximation $A(h)$ of a quantity of interest $A$ given by :
$$
A(h)=A+a_1 h+a_2 h^2+O\left(h^k\right)
$$
The goal of Richardson's method is to eliminate the $h^i$ terms using multiple evaluations $A\left(h_1\right), A\left(h_2\right), \ldots$ Let $\alpha$ be a known and fixed positive constant. Setting $h_i=\alpha^i h$ for some $h>0$, we define :
$$
B(h)=\frac{A(\alpha h)-\alpha A(h)}{1-\alpha}=A+O\left(h^2\right)
$$
Thus, $B$ eliminates the error term $a_1 h$. This process can be repeated to eliminate the remaining terms.

14. Let $A_{i, 0}=A\left(\alpha^i h\right)$ and $A_{1,1}=B(h)$. Show that the expression given by $A_{i, j}=\frac{A_{i, j-1}-\alpha^j A_{i-1, j-1}}{1-\alpha^j}$ removes the first $j$ error terms using $i$ nodes.


Case $j = 0$ is special ($0$ term error is too simple to verify).

Let $i \in ℕ^*$ and $j = 1$:

\begin{align}
A_{i,1} &= \frac{A_{i,0}-\alpha A_{i-1,0}}{1-\alpha} \\
&= \frac{A(\alpha^i h)-\alpha A(\alpha^{i-1} h)}{1-\alpha} \\
&= \frac{A+a_1 \alpha^i h +a_2 (\alpha^i h)^2+O\left(h^2\right)-\alpha (A+a_1 \alpha^{i-1}h +a_2 (\alpha^{i-1} h)^2+O\left(h^2\right))}{1-\alpha} \\
&= \frac{A (1-\alpha) + a_1(\alpha^i h - \alpha\alpha^{i-1}h)+ a_2((\alpha^i h)^2 - \alpha(\alpha^{i-1} h)^2) + O\left(h^2\right)}{1-\alpha} \\
&= A + O\left(h^2\right)
\end{align}

**Conclusion : $A_{i,1}$ removes the first error term using $i$ nodes**

Suppose that $\forall i \in ℕ^*, A_{i,j-1}$ removes the first $j-1$ error terms using $i$ nodes.

Let $i \in ℕ^*$:

Then :

\begin{align}
A_{i,j-1} &= A(\alpha^i h) - a_1 (\alpha^i h)^1 - ... - a_{j-1}(\alpha^i h)^{j-1} \\
A_{i-1,j-1} &= A(\alpha^{i-1} h) - a_1 (\alpha^{i-1} h)^1 - ... - a_{j-1}(\alpha^{i-1} h)^{j-1}
\end{align}

We will now consider the error term with $a_j$ in $A_{i,j}$:

\begin{align}
A_{i,j} &= \frac{A_{i,j-1}-\alpha^j A_{i-1,j-1}}{1-\alpha^j} \\
&= \frac{A + a_j(\alpha^i h)^j-\alpha^j (A + a_j(\alpha^{i-1} h)^j)}{1-\alpha^j} + O\left(h^{j+1}\right) \\
&= \frac{A(1-\alpha^j) + a_j((\alpha^i h)^j-\alpha^j (\alpha^{i-1} h)^j)}{1-\alpha^j} + O\left(h^{j+1}\right) \\
&= A + a_j h^j\frac{(\alpha^i)^j-\alpha^j (\alpha^{i-1})^j}{1-\alpha^j} + O\left(h^{j+1}\right) \\
&= A + a_j h^j\frac{\alpha^{ij}-\alpha^j\alpha^{ji-j}}{1-\alpha^j} + O\left(h^{j+1}\right) \\
&= A + a_j h^j\frac{\alpha^{ij}-\alpha^{ij}}{1-\alpha^j} + O\left(h^{j+1}\right) \\
&= A + O\left(h^{j+1}\right) \\
\end{align}

**Conclusion : $A_{i,j}$ removes the first $j$ error terms using $i$ nodes if $\forall i \in ℕ^*, A_{i,j-1}$ removes the first $j-1$ error terms using $i$ nodes.**


**Conclusion : By recursivity, $A_{i,j}$ removes the $j$ first error terms using $i$ nodes**




15. Richardson's acceleration applied to the trapezoidal rule is called Romberg's integration. The error of the trapezoidal rule is given by :
$$
\widehat{I_t(h)}=\int f+\sum_{k=1}^n a_k h^{2 k}+O\left(h^{2 n+2}\right)
$$
Implement Richardson's method applied to $I_t(h)$.


Before implementing the method, we should notice that :

\begin{align}
\widehat{I_t(h)} &= \int f+\sum_{k=1}^n a_k h^{2 k}+O\left(h^{2 n+2}\right) \\
&= \int f+\sum_{k=1}^n a_k \widetilde{h}^{k}+O\left(\widetilde{h}^{n+1}\right) \\ 
& = \widetilde{A}(\widetilde{h}).
\end{align}

Using the variable change  :
    $\left\{\begin{array}{ll}
        \widetilde{h} = h^2 \\
        \widetilde{A}(x) = A(\sqrt{x})
    \end{array}\right\}$. 

So $\widetilde{A}(\widetilde{h}) = A(h)$. Then :

\begin{align}
\forall i \in \{0,...,n\}, \widetilde{A}_{i,0} &= \widetilde{A}(\alpha^{i}\widetilde{h}) \\
&= A(\sqrt{\alpha^{i}}\sqrt{\widetilde{h}}) \\
&= A(h\sqrt{\alpha^{i}}) \\
&= A(h\alpha^{\frac{i}{2}}) \\
&= A(\frac{b - a}{M}\alpha^{\frac{i}{2}}) \\
&= A(\frac{b - a}{M\alpha^{\frac{-i}{2}}}) \\
&= A(\frac{b - a}{\widetilde{M}}) \\
\end{align}

Using the variable change $\widetilde{M} = M\alpha^{\frac{-i}{2}}$.

So to compute $A_{i,0}$, we need to use $\text{trapezoide\_rule}(f, \widetilde{M}, a, b)$. In the following, we will take $\alpha = \frac{1}{4}$ (which usually gives good result). So $\alpha^{\frac{-i}{2}} = \sqrt{4}^{i} = 2^i$ 

**To conclude, using this particular value of alpha :** $$A_{i,0} = \text{trapezoide\_rule}(f, 2^iM, a, b)$$

Then, we only need to use the recurrency relationship to get $I_t(h)$.

````python
import numpy as np
from scipy.integrate import romberg

alpha = 1/4 # numerator and denominator must be squared integers.


def romberg_integration(f, n, M, a, b):

    # M is the number of intervals
    # n is the level of interpolation

    A = np.zeros((n, n))

    for i in range(n):
        A[i, 0] = trapezoide_rule(f, M * int(1/np.sqrt(alpha)**i), a, b)

        for j in range(1, i+1):
            A[i, j] = (A[i, j-1] - (alpha**j)*A[i-1, j-1]) / (1-alpha**j)
            
    return (A[n-1,n-1])


# Example usage
def g(x):
  return np.exp(-x**2)

a, b = 0, 1
n = 4
M = 4
I = romberg_integration(g, n, M, a, b)
print(f"Romberg integration = {I}")
I2 = romberg(g, a, b)
print("true value : " + str(I2))
````
